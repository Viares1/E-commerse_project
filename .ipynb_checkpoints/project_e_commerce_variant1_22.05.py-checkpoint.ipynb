### Этап 2. Найти количество пользователей, совершивших покупку только один раз.
Для того, чтобы ответить на этот вопрос, надо дать определение пользователю, совершившему покупку. Пользователем, совершившим покупку, будем считать id пользователя (customer_unique_id), который оплатил заказ (статус заказа(order_status) должен быть delivered(доставлен пользователю), shipped(отгружен со склада), created(создан), approved(заказ подтвержден) или processing(в процессе сборки заказа)). Таких оплат у пользователя должно быть ровно 1. При этом если пользователь успешно оплатил один заказ и одна/несколько оплат неуспешны, то будем считать, что у него 1 заказ (так как одна успешная покупка).

```python
# Фильтруем по нужному статусу заказа, группируем по пользователю и подсчитываем тех пользователей, у которых 1 заказ.
sole_order_users_count = full_data_1.query('order_status == "delivered" | order_status == "shipped" | \
                                  order_status == "created" | order_status == "approved" | order_status == "processing"') \
                                  .groupby('customer_unique_id', as_index=False) \
                                  .agg({'customer_id':'count'}) \
                                  .query('customer_id == 1') \
                                  .shape[0]
sole_order_users_count
```

Итого: 91819 пользователей совершивших покупку только один раз

### Этап 3. Узнать количество заказов в месяц в среднем, не доставляющихся по разным причинам, вывести детализацию по причинам.
Чтобы ответить на этот вопрос, надо дать определение недоставленному заказу. Будем считать, что недоставленный заказ - это любой заказ, который пока еще не получил пользователь на момент записи, т.е таковым будет любой заказ, со статусом заказа, отличным от 'delivered' (даже если заказ успешно оплачен или вообще в пути, по факту он не доставлен в этом месяце). Найдем количество недоставленных заказов и сколько месяцев всего сервис занимается доставкой, найдем среднее количество недоставленных заказов в месяц и количество недоставленных заказов в каждый месяц по каждой из причин.

```python
# Переводим даты в нужный формат.
times = ['order_purchase_timestamp', 'order_approved_at', 'order_delivered_carrier_date', 'order_delivered_customer_date', \
'order_estimated_delivery_date']
for i in times:
    full_data_1[i] = pd.to_datetime(full_data_1[i])
```

Для дальнейшего анализа посмотрим на распределение покупок по месяцам и обозначим временные границы для анализа.

```python
# Приводим даты к началу каждого месяца, создаем колонку.
full_data_1['months'] = full_data_1['order_purchase_timestamp'].to_numpy().astype('datetime64[M]')

# Считаем количество покупок по месяцам
full_data_1.groupby('months').agg({'order_id':'count'})
```

Видим аномально-низкие показатели количества ордеров в периоды 2016-09-01 - 2016-12-01 включительно и 2018-09-01 - 2018-10-01 включительно. Чтобы получить более объективные результаты анализа, не учитываем эти периоды в нашем дальнейшем анализе.

```python
# Создаем таблицу с причиной, по которой заказ не доставлен и количеством таких заказов в каждый из месяцев в выбранном периоде.
not_delivered_by_month = full_data_1.query('months >= "2017-01-01" & months < "2018-09-01"') \
                                    .query('order_status != "delivered"') \
                                    .groupby(['months', 'order_status'], as_index=False) \
                                    .agg({'order_id':'count'}) \
                                    .pivot(columns='order_status', index='months', values='order_id').fillna(0).reset_index()
not_delivered_by_month
```

```python
# Находим среднее количество недоставленных заказов в месяц.
not_delivered_avg = (not_delivered_by_month.canceled.sum() + not_delivered_by_month.unavailable.sum() + \
                     not_delivered_by_month.approved.sum() + not_delivered_by_month.created.sum() + \
                     not_delivered_by_month.invoiced.sum() + not_delivered_by_month.processing.sum() + \
                     not_delivered_by_month.shipped.sum()) / not_delivered_by_month.months.count()

# Добавляем колонку с этим значением в нашу таблицу, округляем до целого.
not_delivered_by_month['monthly_avg'] = round(not_delivered_avg)

# Переименовываем наши столбцы для удобства.
not_delivered_by_month = not_delivered_by_month.rename(columns={'months'     :'Месяц',
                                                                'approved'   :'Подтверждён',
                                                                'canceled'   :'Отменён',
                                                                'created'    :'Создан',
                                                                'invoiced'   :'Выставлен счёт',
                                                                'processing' :'В процессе сборки заказа',
                                                                'shipped'    :'Отгружен со склада',
                                                                'unavailable':'Недоступен',
                                                                'monthly_avg':'Недоставленных заказов в месяц, среднее'})
not_delivered_by_month
```

```python
# Изобразим все результаты на общем графике.
sns.lineplot(data=not_delivered_by_month, x='Месяц', y='Отменён', label='Отменён')
sns.lineplot(data=not_delivered_by_month, x='Месяц', y='Выставлен счёт', label='Выставлен счёт')
sns.lineplot(data=not_delivered_by_month, x='Месяц', y='В процессе сборки заказа', label='В процессе сборки заказа')
sns.lineplot(data=not_delivered_by_month, x='Месяц', y='Отгружен со склада', label='Отгружен со склада')
sns.lineplot(data=not_delivered_by_month, x='Месяц', y='Недоступен', label='Недоступен')
sns.lineplot(data=not_delivered_by_month, x='Месяц', y='Недоставленных заказов в месяц, среднее', \
    label='Недоставленных заказов в месяц, среднее').set(title='Количество недоставленных заказов по каждой из причин.')
plt.legend()
plt.legend(loc='upper left')
plt.ylabel('Количество заказов')
plt.show()
```

### Этап 4. По каждому товару определить, в какой день недели товар чаще всего покупается.
Чтобы ответить на этот вопрос, надо для каждого продукта определить день/дни недели когда его покупают больше всего.

```python
# Объединим датафрейм с информацией о заказах и датафрейм с информацией о товарах.
full_data_2 = orders.merge(orders_items, how = 'left', on = 'order_id')
full_data_2.head()
```

```python
# Переведем даты в нужный формат.
times = ['order_purchase_timestamp', 'order_approved_at', 'order_delivered_carrier_date', 'order_delivered_customer_date', \
'order_estimated_delivery_date', 'shipping_limit_date']
for i in times:
    full_data_2[i] = pd.to_datetime(full_data_2[i])
```

```python
# Добавим в таблицу номер дня недели, когда был куплен товар.
full_data_2['weekday'] = full_data_2['order_purchase_timestamp'].dt.isocalendar().day

# Реализуем функцию, которая в качестве аргумента принимает серию с номером дня недели и возвращает список с соответствующими
# номерам названиями дней недели.
def day_of_week(series):
    my_dict = {1 : 'Понедельник', 2 : 'Вторник', 3 : 'Среда', 4 : 'Четверг', 5 : 'Пятница', 6 : 'Суббота', 7 : 'Воскресение'}
    my_lst = []
    for i in series:
        my_lst.append(my_dict[i])
    return my_lst

# Создадим колонку с названиями дней недели.
full_data_2['weekday'] = day_of_week(full_data_2['weekday'])
```

```python
# Найдем дни недели, когда каждый товар купили максимальное количество раз.
count_per_day = full_data_2.query('order_status == "delivered" | order_status == "shipped" | \
                           order_status == "created" | order_status == "approved" | order_status == "processing"') \
                           .groupby(['product_id', 'weekday'], as_index=False) \
                           .agg({'order_id':'count'})

# Найдем количество максимальных покупок в день для каждого товара.
max_for_product = full_data_2.query('order_status == "delivered" | order_status == "shipped" | \
                             order_status == "created" | order_status == "approved" | order_status == "processing"') \
                             .groupby(['product_id', 'weekday'], as_index=False) \
                             .agg({'order_id':'count'}) \
                             .groupby('product_id', as_index=False) \
                             .agg({'order_id':'max'})

# Объединим 2 датафрейма, созданных выше.
days_max_buys = count_per_day.merge(max_for_product, how='left', on='product_id')

# Оставим только записи с днями, где количество покупок равно максимальному количеству покупок каждого продукта (т.е. у каждого
# товара свой максимум покупок).
days_max_buys = days_max_buys.query('order_id_x == order_id_y')

# Для каждого товара создаем строку, где перечисляются все дни, когда его покупали чаще всего.
days_max_buys = days_max_buys.groupby('product_id', as_index=False) \
                             .agg({'weekday': lambda x: ",".join(x)})
days_max_buys
```

```python
# Подсчитаем количество товаров, когда их чаще всего покупали для каждого дня. Товары, у которых несколько таких дней объединим 
# в одну группу - "Несколько топ-дней".
days_max_buys = days_max_buys.groupby('weekday', as_index=False) \
                             .agg({'product_id':'count'}) \
                             .sort_values('product_id', ascending=False)

# Создадим колонку с количеством дней в строке.
days_max_buys['days_num'] = days_max_buys.weekday.apply(lambda x: len(x.split(',')))
days_max_buys
```

```python
# Отсортируем строки: включаем все строки с количеством дней больше 1.
days_max_buys_several = days_max_buys.query('days_num > 1')

# Подсчитаем количество строк, подходящих под описание "Несколько топ-дней".
several_days_products_count = days_max_buys_several.product_id.sum()

# Создадим новую строку с группой товаров, у которых несколько топ-дней и количеством таких товаров.
new_row = {'weekday':'Несколько топ-дней', 'product_id':several_days_products_count}

# Берем строки, где 1 топ-день.
days_max_buys = days_max_buys.query('days_num == 1')

# Собираем всю информацию в единую таблицу, приводим к нормальному виду.
days_max_buys = days_max_buys.drop(columns=['days_num'])
days_max_buys = days_max_buys.append(new_row, ignore_index=True) \
                             .rename(columns={'weekday':'Топ день по покупкам товара',
                                              'product_id':'Количество товаров'}) \
                             .sort_values('Количество товаров', ascending=False)
days_max_buys
```

```python
# Изобразим результаты на графике.
sns.barplot(data=days_max_buys, x="Топ день по покупкам товара", y="Количество товаров") \
    .set(title='Количество товаров по дням, когда их чаще всего покупали.');
plt.xticks(rotation=70);
```

Итого:
- 7037 товаров, у которых несколько дней, в которые их покупали чаще всего.
- Остальные товары распределились по конкретным дням, дни расположились в следующем порядке: понедельник - 4275 товаров, вторник - 4136 товаров, среда - 4087 товаров, четверг - 3786 товаров, пятница - 3673 товара, воскресение - 2973 товара, суббота - 2649 товаров.

### Этап 5. Посчитать среднее количество покупок в неделю у каждого из пользователей по месяцам.
Чтобы посчитать эти значения, надо найти количество покупок каждым пользователем в каждый месяц и разделить на количество недель в конкретном месяце. Так же найдем количество недель в каждом месяце.

```python
# Рассчитаю количество недель для всех длин месяца (в феврале 2017г. и 2018г. по 28 дней). Возьму уже имеющийся датафрейм.
full_data_3 = full_data_1.copy()

# Реализую функцию, которая в качестве аргумента принимает серию с датой совершения заказа и возвращает список с количеством
# недель в месяце, который фигурирует в каждой из дат.
def coeff(series):
    my_lst = []
    month_length = {1 : 4.43, 3 : 4.43, 5 : 4.43, 7 : 4.43, 8 : 4.43, 10 : 4.43, 12 : 4.43, 4 : 4.29, \
    6 : 4.29, 9 : 4.29, 11 : 4.29, 2 : 4}
    for i in series:
        my_lst.append(month_length[i.month])
    return my_lst

# Создаю столбец с количеством недель в месяце.
full_data_3['coefficient'] = coeff(full_data_3['order_purchase_timestamp'])
```

```python
# Группируем по месяцу и пользователю и подсчитываем количество заказов для такой связки.
orders_per_month_week_per_user = full_data_3.query('order_status == "delivered" | order_status == "shipped" | \
                                   order_status == "created" | order_status == "approved" | order_status == "processing"') \
                                            .query('months >= "2017-01-01" & months < "2018-09-01"') \
                                            .groupby(['months', 'customer_unique_id', 'coefficient'], as_index=False) \
                                            .agg({'order_id':'count'})

# Считаем среднее количество заказов, учитывая наш коэффициент, округляем.
orders_per_month_week_per_user['avg_per_week'] = round(orders_per_month_week_per_user.order_id / \
                                                       orders_per_month_week_per_user.coefficient, 3)

# Сортируем по количеству заказов и выводим.
orders_per_month_week_per_user = orders_per_month_week_per_user.sort_values('avg_per_week', ascending=False) \
                                                               .rename(columns={'order_id':'orders_count'})
orders_per_month_week_per_user
```

Итого: мы нашли cреднее число покупок в неделю по месяцам для каждого конкретного пользователя. Значения, когда пользователь не совершил ни одной покупки за месяц, отсутствуют. Ответ - колонка 'avg_per_week'.

### Этап 6. Провести когортный анализ пользователей. В период с января по декабрь выявить когорту с самым высоким retention на 3й месяц.
Для того, чтобы провести когортный анализ, надо найти месяц первой покупки для всех пользователей и найти retention rate для разных когорт покупателей.

```python
# Копируем имеющийся датафрейм и ищем месяц первой покупки для каждого пользователя.
full_data_4 = full_data_1.copy()
first_month_per_user = full_data_4.query('order_status == "delivered" | order_status == "shipped" | \
                                  order_status == "created" | order_status == "approved" | order_status == "processing"') \
                                  .query('months >= "2017-01-01" & months < "2018-04-01"') \
                                  .groupby('customer_unique_id') \
                                  .agg({'months':'min'})

# Создаем датафрейм с записями, подходящими для наших условий (временных и относительно статуса заказа).
cohort = full_data_4.query('order_status == "delivered" | order_status == "shipped" | \
                    order_status == "created" | order_status == "approved" | order_status == "processing"') \
                    .query('months >= "2017-01-01" & months < "2018-04-01"')

# Соединяем полученные выше 2 датафрейма.
cohort = cohort.merge(first_month_per_user, how='left', on='customer_unique_id') \
               .rename(columns={'months_x':'months', 'months_y':'first_buy'})
cohort.head()
```

Найдем retention rate для пользователей, сделавших первую покупку в период с января по декабрь. Retention rate 3-его месяца -
количество пользователей, сделавших повторный заказ спустя 3 месяца.

```python
# Реализуем функцию, которая в качестве аргумента принимает датафрейм и возвращает список значений retantion 3-его месяца 
# для каждого рассматриваемого месяца. Эта функция понадобится нам в дальнейшем.
def retention(my_cohort):
    my_lst = []
    for i in my_cohort.columns[:-3]:
        my_lst.append(round(my_cohort[i][i.month + 2] / my_cohort[i][i.month - 1], 4))
    return my_lst
```

Найдем retention для пользователей, получивших заказ вовремя (теоретическое время доставки, указанное продавцом должно быть 
больше или равно фактического времени доставки).

```python
# Приводим дату доставки к нужному типу.
cohort['order_delivered_customer_date'] = cohort['order_delivered_customer_date'].to_numpy().astype('datetime64[D]')

# Фильтруем покупателей по вышеуказанному условию и создаем pivot-таблицу, в которой удобно рассчитывать retention.
final_cohort1 = cohort.query('order_delivered_customer_date <= order_estimated_delivery_date') \
                      .groupby(['months', 'first_buy'], as_index=False) \
                      .agg({'customer_unique_id':'nunique'}) \
                      .pivot(columns='first_buy', index='months', values='customer_unique_id').fillna(0)
final_cohort1
```

```python
# Рассчитываем retantion для выведенной когорты покупателей.
print(retention(final_cohort1))
```

Максимальный retantion 3-его месяца для пользователей, получивших заказ вовремя, наблюдается для пользователей, сделавших
первую покупку в марте и составляет 0.42%.

Разобьем пользователей на когорты по штатам. Т.к. количество повторных покупок небольшое, возьмем только первые 3 штата и 
подсчитаем retantion.

```python
cohort.customer_state.value_counts().head(3)
```

```python
# Разбиваем пользователей на когорты и для каждой когорты подсчитываем значения retention.
for i in ['SP', 'RJ', 'MG']:
    final_cohort2 = cohort.query('customer_state == @i') \
                          .groupby(['months', 'first_buy'], as_index=False) \
                          .agg({'customer_unique_id':'nunique'}) \
                          .pivot(columns='first_buy', index='months', values='customer_unique_id').fillna(0)
    print(retention(final_cohort2))
```

Наибольший retantion 3-его месяца среди когорт по штатам SP, RJ, MG у пользователей, сделавших первую покупку в марте из штата
SP, составляет 0.72%.

Так же возьмем когорту пользователей из города Sao Paulo.

```python
# Берем когорту и рассчитываем значения retention.
final_cohort3 = cohort.query('customer_city == "sao paulo"') \
                      .groupby(['months', 'first_buy'], as_index=False) \
                      .agg({'customer_unique_id':'nunique'}) \
                      .pivot(columns='first_buy', index='months', values='customer_unique_id').fillna(0)
print(retention(final_cohort3))
```

Наибольший retantion 3-его месяца составляет 0.82% для пользователей, совершивших первую покупку в марте. Sao Paulo находится в штате SP.

Выделим когорты пользователей, делающих дорогие и дешевые заказы. Дорогим заказом будем считать заказ, стоимость которого выше
средней стоимости заказа, дешевым, соответственно, тот, что дешевле средней стоимости заказа.

```python
# Берем когорту и рассчитываем значения retention.
final_cohort4 = cohort.query('order_id in @expensive_orders') \
                      .groupby(['months', 'first_buy'], as_index=False) \
                      .agg({'customer_unique_id':'nunique'}) \
                      .pivot(columns='first_buy', index='months', values='customer_unique_id').fillna(0)
print(retention(final_cohort4))
```

Наибольший retention 3-его месяца для дешевых заказов для покупателей, сделавших первый заказ в марте, составляет 0.44%.

Выделим когорты пользователей, делающих тяжелые и легкие заказы. Тяжелым заказом будем считать заказ, вес которого выше среднего веса заказа, легким, соответственно, тот, что легче среднего веса заказа.

```python
# Берем когорту и рассчитываем значения retention.
final_cohort6 = cohort.query('order_id in @heavy_orders') \
                      .groupby(['months', 'first_buy'], as_index=False) \
                      .agg({'customer_unique_id':'nunique'}) \
                      .pivot(columns='first_buy', index='months', values='customer_unique_id').fillna(0)
print(retention(final_cohort6))
```

Наибольший retention 3-его месяца для тяжелых заказов для покупателей, сделавших первый заказ в мае, составляет 0.43%.

Наибольший retention 3-его месяца для легких заказов для покупателей, сделавших первый заказ в марте и июне, составляет 0.48%.

### Этап 7. Провести кластеризацию пользователей по RFM-сегментам.
Для того, чтобы провести кластеризацию пользователей по RFM-сегментам, надо отобрать подходящие под условия (временные и статус заказа) записи, распределить пользователей на сегменты, взависимости от их значений метрик и затем распределить сегменты по кластерам. В качестве метрик для RFM-сегментации возьму следующие показатели: recency -  время от последней покупки пользователя до текущей даты, frequency - суммарное количество покупок у пользователя за всё время, monetary - сумма покупок за всё время.

```python
# Делаем копию уже имеющегося датафрейма. Находим месяц первой покупки для каждого покупателя.
full_data_5 = full_data_1.copy()
rfm_data_temp = full_data_5.query('order_status == "delivered" | order_status == "shipped" | \
                           order_status == "created" | order_status == "approved" | order_status == "processing"') \
                           .query('months >= "2017-01-01" & months < "2018-09-01"') \
                           .groupby('customer_unique_id') \
                           .agg({'months':'min'})

# Создаем датафрейм с подходящими по условиям записями.
rfm_data = full_data_5.query('order_status == "delivered" | order_status == "shipped" | \
                      order_status == "created" | order_status == "approved" | order_status == "processing"') \
                      .query('months >= "2017-01-01" & months < "2018-09-01"')

# Соединяем 2 датафрейма, созданных выше.
rfm_data = rfm_data.merge(rfm_data_temp, how='left', on='customer_unique_id') \
                   .rename(columns={'months_x':'months', 'months_y':'first_buy'})

# Добавляем информацию об товарах в заказах.
rfm_data = rfm_data.merge(orders_items, how='left', on='order_id')

# Находим время последней записи в датафрейме.
rfm_data['now'] = rfm_data['order_purchase_timestamp'].max()

# Добавляем столбец с последней покупкой для каждого пользователя.
rfm_data = rfm_data.merge(rfm_data.groupby('customer_unique_id', as_index=False) \
                                  .agg({'order_purchase_timestamp':'max'}), how='left', on='customer_unique_id')

# Рассчитываем интервал между временем покупки в строке и временем последней записи в датафрейме.
rfm_data['recency'] = rfm_data['now'] - rfm_data['order_purchase_timestamp_y']

# Удаляем ненужные колонки.
rfm_data = rfm_data.drop(columns=['now', 'order_purchase_timestamp_y', 'months'])

# Переводим интервалы, найденные выше, в дни.
rfm_data['recency'] = rfm_data['recency'].dt.days

# Подсчитываем метрики R, F, M для каждого пользователя.
rfm_data = rfm_data.groupby('customer_unique_id', as_index=False) \
    .agg({'recency':'max', 'order_id':'count', 'price':'sum'}) \
    .sort_values('order_id', ascending=False) \
    .rename(columns={'order_id':'frequency', 'price':'monetary'})
rfm_data.head()
```

Разбиваю recency, monetary метрики на 3 подгруппы, по 0.33 и 0.66 квантилю. В случае с frequency - практически все 
пользователи за все время совершили 1 покупку и медианное значение практически для любого квантиля равно 1, так что я разбил 
эту метрику на следующие 3 подгруппы: те, кто сделал 1 покупку, кто сделал 2 покупки (медианное значение, если отбросить всех,
кто сделал 1 покупку)  и все остальные. Таким образом подгруппы по метрикам следующие:
- recency: 1 - пользователь совершил последнюю покупку более 285 дней назад, 2 - пользователь совершил последнюю покупку более 147, но менее или равно 285 дней назад, 3 - пользователь совершил последнюю покупку менее или равно 147 дней назад;
- frequency: 1 - пользователь совершил 1 покупку за все время, 2 - пользователь совершил 2 покупки за все время, 3 - пользователь совершил более 2 покупок за все время;
- monetary: 1 - пользователь потратил менее или равно 57.99 долларов за все время, 2 - пользователь потратил более 57.99 долларов, но менее или равно 124.9 долларов за все время, 3 - пользователь потратил более 124.9 долларов за все время.

```python
# Создаем словарь, ключи в котором это названия метрик (кроме frequency), а значения - это словари, ключи в которых это квантили,
# а значения - соответствующие этим квантилям значения метрик.
quantiles = rfm_data[['recency', 'monetary']].quantile([0.33, 0.66]).to_dict()

# Рассчитываю медиану метрики frequency, если убрать все строки, значения frequency в которых = 1.
rfm_freq_med = rfm_data.query('frequency > 1') \
                       .frequency.median()

# Добавляю значения для метрики frequency в словарь с квантилями.
quantiles['frequency'] = {0.33: 1, 0.66: rfm_freq_med}
quantiles
```

```python
# Реализуем функции для распределения покупателей по сегментам. Функциям в качестве аргумента подается значения метрик (в 
# случае со второй функцией дополнительно подается названием метрики), а возвращается номер сегмента, которому принадлежит
# пользователь. У метрики R обратная зависимость, поэтому нужна отдельная функция.
def recency(x):
    if x <= quantiles['recency'][0.33]:
        return 3
    elif quantiles['recency'][0.33] < x <= quantiles['recency'][0.66]:
        return 2
    else:
        return 1
    
def frequency_monetary(x, segment):
    if x <= quantiles[segment][0.33]:
        return 1
    elif quantiles[segment][0.33] < x <= quantiles[segment][0.66]:
        return 2
    else:
        return 3

# Создаем колонки R, F, M и записываем туда ранг пользователя, в зависимости от значения его метрики.
rfm_data['R'] = rfm_data['recency'].apply(lambda x: recency(x))
rfm_data['F'] = rfm_data['frequency'].apply(lambda x: frequency_monetary(x, 'frequency'))
rfm_data['M'] = rfm_data['monetary'].apply(lambda x: frequency_monetary(x, 'monetary'))

# Создаем колонку с RFM-сегментом пользователя.
rfm_data['RFM'] = rfm_data['R'].astype('str') + rfm_data['F'].astype('str') + rfm_data['M'].astype('str')
```

Разбиваем сегментированных покупателей на кластеры взависимости от RFM показателя:
- Тестер - очень давно сделал 1-2 маленьких заказа;
- Забросивший покупатель - сделал один/несколько небольших/средних заказов, но давно не проявлял активность;
- Забросивший топ-покупатель - сделал один/несколько заказов на большую сумму, но давно не проявлял активность;
- Потенциальный покупатель - относительно недавно сделал 1-2 маленьких заказа, возможно возобновит активность;
- Разовый покупатель - относительно давно сделал средний заказ;
- Разовый топ-покупатель - относительно давно сделал большой заказ;
- Покупатель - сделал один/несколько небольших/средних заказов, последний относительно недавно, возможно возобновит активность;
- Топ-покупатель - сделал один/несколько заказов на большую сумму, относительно недавно проявлял активность;
- Активный потенциальный покупатель - сделал 1-2 маленьких заказа, последний из которых недавно;
- Активный покупатель - сделал один/несколько небольших/средних заказов, последний из которых недавно;
- Активный топ-покупатель - сделал один/несколько заказов на большую сумму, последний из которых недавно;
- Лучший покупатель - регулярно делает заказы, суммарно на большую сумму.

```python
# Создаем словарь, где ключи это сегменты покупателей, а значения - соответствующие сегментам кластеры.
segments = {'111':'Тестер',                               '121':'Тестер',                            '112':'Разовый покупатель', 
            '122':'Забросивший покупатель',               '123':'Забросивший топ-покупатель',        '132':'Забросивший покупатель',
            '133':'Забросивший топ-покупатель',           '131':'Забросивший покупатель',            '113':'Разовый топ-покупатель',
            '211':'Потенциальный покупатель',             '221':'Потенциальный покупатель',          '231':'Покупатель',
            '212':'Разовый покупатель',                   '222':'Покупатель',                        '232':'Покупатель',
            '223':'Топ-покупатель',                       '233':'Топ-покупатель',                    '213':'Разовый топ-покупатель',
            '311':'Активный потенциальный покупатель',    '321':'Активный потенциальный покупатель', '312':'Активный покупатель',
            '322':'Активный покупатель',                  '323':'Активный топ-покупатель',           '332':'Активный покупатель',
            '333':'Лучший покупатель',                    '331':'Активный покупатель',               '313':'Активный топ-покупатель'}

# Создаем колонку с названием кластера для каждого покупателя.
rfm_data['RFM_claster'] = rfm_data['RFM'].apply(lambda x: segments[x])
rfm_data.head()
```

Найдем количество покупателей для каждого кластера и изобразим значения на графике. Так же найдем распределение метрик R, F, M по подгруппам и изобразим на графиках.

```python
# Находим количество покупателей в каждом из кластеров.
rfm_categories = rfm_data.groupby('RFM_claster', as_index=False) \
                         .agg({'customer_unique_id':'count'}) \
                         .sort_values('customer_unique_id', ascending=False) \
                         .rename(columns={'customer_unique_id':'Количество покупателей',
                                          'RFM_claster':'Кластер'})
rfm_categories
```

```python
# Изображаем на графике количество покупателей по кластерам.
sns.barplot(data=rfm_categories, x="Кластер", y="Количество покупателей") \
    .set(title='Количество покупателей по кластерам, образованным в результате RFM сегментации.');
plt.xticks(rotation=80);
```

```python
# Находим количество покупателей в каждой подгруппе для каждой метрики.
rfm_data_R = rfm_data.groupby('R', as_index=False) \
                     .agg({'customer_unique_id':'count'}) \
                     .rename(columns={'R':'Requency', 'customer_unique_id':'Количество пользователей в подгруппе'})
rfm_data_F = rfm_data.groupby('F', as_index=False) \
                     .agg({'customer_unique_id':'count'}) \
                     .rename(columns={'F':'Frequency', 'customer_unique_id':'Количество пользователей в подгруппе'})
rfm_data_M = rfm_data.groupby('M', as_index=False) \
                     .agg({'customer_unique_id':'count'}) \
                     .rename(columns={'M':'Monetary', 'customer_unique_id':'Количество пользователей в подгруппе'})
```

```python
# Изображаем на графике количество покупателей по подгруппам в метрике Requency.
sns.barplot(data=rfm_data_R, x="Requency", y="Количество пользователей в подгруппе") \
    .set(title='Количество покупателей по подгруппам в метрике Requency.');
```

```python
# Изображаем на графике количество покупателей по подгруппам в метрике Frequency.
sns.barplot(data=rfm_data_F, x="Frequency", y="Количество пользователей в подгруппе") \
    .set(title='Количество покупателей по подгруппам в метрике Frequency.');
```

```python
# Изображаем на графике количество покупателей по подгруппам в метрике Monetary.
sns.barplot(data=rfm_data_M, x="Monetary", y="Количество пользователей в подгруппе") \
    .set(title='Количество покупателей по подгруппам в метрике Monetary.');
```
